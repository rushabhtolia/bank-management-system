```python
import sqlite3
from IPython.display import clear_output

conn = sqlite3.connect('Bank Server1.db')
c = conn.cursor()


print("Welcome to ABC Bank")
print("*"*20)

print("Enter User ID: ")
iuser=int(input(""))
print("Enter PIN: ")
ipin=int(input(""))

c.execute("SELECT userid FROM info WHERE userid = ? ",(iuser,))
testid=c.fetchone()
c.execute("SELECT pin FROM info WHERE userid = ? ",(iuser,))
testpin=c.fetchone()

if (testid == None or testpin==None):
    clear_output()
    print("Invalid UserID or PIN")
    exit()



elif( iuser==testid[0] and ipin == testpin[0]):
    c.execute("SELECT balance FROM info WHERE userid=?",(iuser,))
    balance=c.fetchone()
    clear_output()
    print("Select an Option")
    print(" ")
    print("1. Withdrawal")
    print("2. Deposit")
    print("3. Balance Check")
    print("4. Exit")
    num=input("")
    
    while(num!='4'):
        
        if(num == '1'):
            clear_output()
            print("Enter amount to be Withdrawn")
            w_amt=int(input(""))
            
            if(w_amt <= balance[0]):
                testbal=balance[0]-w_amt;
                c.execute("UPDATE info SET balance={} WHERE userid={}".format( testbal, iuser))
                conn.commit()
                print("Amount has been withdrawn")
                c.execute("SELECT balance FROM info WHERE userid=?",(iuser,))
                newbalance=c.fetchone()
                print("Updated balance: ",newbalance[0])
                print(" ")
                print("Enter next input: ")
                print(" ")
                print("1. Withdrawal")
                print("2. Deposit")
                print("3. Balance Check")
                print("4. Exit")
                num=input("")
            else:
                print("Insufficient Balance")
                print(" ")
                print("Enter next input: ")
                print(" ")
                print("1. Withdrawal")
                print("2. Deposit")
                print("3. Balance Check")
                print("4. Exit")
                num=input("")
        
        elif(num == '2'):
            clear_output()
            print("Enter amount to be Deposited")
            d_amt=int(input(""))
            clear_output()
            c.execute("SELECT balance FROM info WHERE userid=?",(iuser,))
            balance=c.fetchone()
            testbal=balance[0]+d_amt;
            c.execute("UPDATE info SET balance={} WHERE userid={}".format( testbal, iuser))
            conn.commit()
            print("Amount has been Deposited")
            c.execute("SELECT balance FROM info WHERE userid=?",(iuser,))
            newbalance=c.fetchone()
            print("Updated balance: ",newbalance[0])
            
            
            print(" ")
            print("Enter next input: ")
            print(" ")
            print("1. Withdrawal")
            print("2. Deposit")
            print("3. Balance Check")
            print("4. Exit")
            num=input("")
        
        elif(num == '3'):
            clear_output()
            print("Your current balance is: ")
            c.execute("SELECT balance FROM info WHERE userid=?",(iuser,))
            balance=c.fetchone()
            print(balance[0])
            print(" ")
            print("Enter next input: ")
            print(" ")
            print("1. Withdrawal")
            print("2. Deposit")
            print("3. Balance Check")
            print("4. Exit")
            num=input("")
        
        
        else:
            clear_output()
            print("Try Again")
            num=input("")
    clear_output()
    print("Thank You")

 

```
